# TP3 : Réseaux sémantiques

## Exercice 1

```python
import networkx as nx
import matplotlib.pyplot as plt

# Création du graphe orienté
G = nx.DiGraph()

# Ajouter les concepts
G.add_node("Animal")
G.add_node("Mammifère")
G.add_node("Poisson")
G.add_node("Dauphin")
G.add_node("Peut nager")

# Ajouter les relations "est-un"
G.add_edge("Mammifère", "Animal", relation="est-un")
G.add_edge("Poisson", "Animal", relation="est-un")
G.add_edge("Dauphin", "Mammifère", relation="est-un")

# Ajouter les propriétés explicitement pour les animaux qui vivent dans l'eau
G.add_edge("Poisson", "Peut nager", relation="propriété")
G.add_edge("Dauphin", "Peut nager", relation="propriété")

# Position des nœuds pour affichage clair
pos = {
    "Animal": (0, 3),
    "Mammifère": (-1, 2),
    "Poisson": (1, 2),
    "Dauphin": (-1, 1),
    "Peut nager": (0, 0)
}

# Dessiner les nœuds
nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=2000)

# Dessiner les arêtes
nx.draw_networkx_edges(G, pos, arrowstyle='->', arrowsize=20)

# Dessiner les labels des nœuds
nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')

# Dessiner les labels des relations
edge_labels = nx.get_edge_attributes(G, 'relation')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red', font_size=9)

# Afficher le graphe
plt.axis('off')
plt.title("Réseau sémantique")
plt.show()
```

## Exercice 3

```python
# Création du graphe orienté
G = nx.DiGraph()

# Ajouter les concepts / nœuds
G.add_node("Personne")
G.add_node("Étudiant")
G.add_node("Enseignant")
G.add_node("Professeur")
G.add_node("Chercheur")
G.add_node("Domaine")
G.add_node("Cours")
G.add_node("Âge (numérique)")

# Ajouter les relations "est-un"
G.add_edge("Étudiant", "Personne", relation="est-un")
G.add_edge("Enseignant", "Personne", relation="est-un")
G.add_edge("Professeur", "Enseignant", relation="est-un")
G.add_edge("Professeur", "Chercheur", relation="est-un")

# Ajouter les propriétés
G.add_edge("Personne", "Âge (numérique)", relation="possède")
G.add_edge("Étudiant", "Domaine", relation="étudie")
G.add_edge("Enseignant", "Domaine", relation="enseigne")

# Ajouter la contrainte d'égalité entre domaine enseigné et étudié pour le même cours
G.add_edge("Cours", "Domaine", relation="domaine étudié = domaine enseigné")

# Position des nœuds pour affichage clair
pos = {
    "Personne": (0, 3),
    "Étudiant": (-2, 2),
    "Enseignant": (2, 2),
    "Professeur": (2, 1),
    "Chercheur": (4, 1),
    "Domaine": (0, 1),
    "Cours": (0, 0),
    "Âge (numérique)": (-3, 3)
}

# Dessiner les nœuds
nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=2000)

# Dessiner les arêtes
nx.draw_networkx_edges(G, pos, arrowstyle='->', arrowsize=20)

# Dessiner les labels des nœuds
nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')

# Dessiner les labels des relations
edge_labels = nx.get_edge_attributes(G, 'relation')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red', font_size=9)

# Afficher le graphe
plt.axis('off')
plt.title("Réseau sémantique")
plt.show()
```

## Exercice 4

```python
G = nx.DiGraph()
nodes = ["ÊtreVivant", "Animal", "Chien", "Chat", "Os", "Lait", "Nourriture"]
G.add_nodes_from(nodes)

G.add_edge("Chien", "Animal", relation="est-un")
G.add_edge("Chat", "Animal", relation="est-un")
G.add_edge("Animal", "ÊtreVivant", relation="est-un")
G.add_edge("Chien", "Os", relation="lié-à")
G.add_edge("Chat", "Lait", relation="lié-à")
G.add_edge("Os", "Nourriture", relation="type")

# --- Position manuelle pour affichage vertical ---
pos = {
    "ÊtreVivant": (0, 4),
    "Animal": (0, 3),
    "Chien": (-1, 2),
    "Chat": (1, 2),
    "Os": (-1, 1),
    "Lait": (1, 1),
    "Nourriture": (0, 0)
}

plt.figure(figsize=(6, 8))
nx.draw(G, pos, with_labels=True,
        node_color="lightblue", node_size=2500,
        font_size=10, font_weight="bold", arrowsize=20)

edge_labels = nx.get_edge_attributes(G, "relation")
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color="red")

plt.title("Réseau sémantique")
plt.axis("off")
plt.show()
```

## Exercice 7

```python
import networkx as nx
import matplotlib.pyplot as plt

# Création du graphe
G = nx.DiGraph()

# Ajouter les nœuds
nodes = ["Maladie", "MaladieRespiratoire", "Grippe", "Pneumonie",
         "Fièvre", "Toux", "DouleurPoitrine"]
G.add_nodes_from(nodes)

# Ajouter les relations ISA (est-un)
G.add_edge("MaladieRespiratoire", "Maladie", relation="isa")
G.add_edge("Grippe", "MaladieRespiratoire", relation="isa")
G.add_edge("Pneumonie", "MaladieRespiratoire", relation="isa")

# Ajouter les relations hasSymptom (a pour symptôme)
G.add_edge("Grippe", "Fièvre", relation="hasSymptom")
G.add_edge("Grippe", "Toux", relation="hasSymptom")
G.add_edge("Pneumonie", "Fièvre", relation="hasSymptom")
G.add_edge("Pneumonie", "DouleurPoitrine", relation="hasSymptom")

# Ajouter les relations symptom (symptôme de)
G.add_edge("Fièvre", "Maladie", relation="symptôme de")
G.add_edge("Toux", "MaladieRespiratoire", relation="symptôme de")

# Positionnement des nœuds
pos = {
    "Maladie": (0, 4),
    "MaladieRespiratoire": (0, 3),
    "Grippe": (-1, 2),
    "Pneumonie": (1, 2),
    "Fièvre": (0, 1),
    "Toux": (-2, 1),
    "DouleurPoitrine": (2, 1)
}

# Dessiner les nœuds
nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=2000)

# Dessiner les arêtes
nx.draw_networkx_edges(G, pos, arrowstyle='->', arrowsize=20)

# Dessiner les labels des nœuds
nx.draw_networkx_labels(G, pos, font_size=9, font_weight='bold')

# Dessiner les labels des relations
edge_labels = nx.get_edge_attributes(G, 'relation')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red', font_size=8)

plt.title("Réseau sémantique", fontsize=12, fontweight='bold')
plt.axis('off')
plt.show()
```
